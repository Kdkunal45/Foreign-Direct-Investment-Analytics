**COLUMNS DETAILS:**

There are two types of columns:

1.   First columns is the "Sector"Column in which there are different 63 sector that have receved FDI from 200-01 to 2016-17.
2.   Other cokumns are year wise columns in which we can see how much different sector receved investment from 2000-01 to 2016-17.



```python
Year =['2000-01','2001-02','2002-03','2003-04','2004-05','2005-06','2006-07','2007-08','2008-09','2009-10','2010-11','2011-12','2012-13','2013-14','2014-15','2015-16','2016-17']
Sectors =['Sector']
```

```python
# Extracting the Detail Information:
df.info()
```

```python
# Checking the null values :
df.isnull().sum()
```

```python
#Creating Average Exchange Rate List Reference (Rbt website)

Rates =[45.68,47.69,48.39,45.95,44.93,44.27,45.24,40.26,45.99, 47.44,45.56,47.92,54.40,60.50,61.14,65.46,67.07]
```

**Converting American '$' to Indian 'Rupees':**

```python

#Creating a function to Convert FDI's value from USD to INR
def multiply_columns(df, col_list, num):
    for col in col_list:
        df[col]= df[col] * Rates [col_list.index(col)]/10
    return df
```

```python
FDI_InUSD = df.copy()

# Assuming Year and Rates are defined elsewhere
FDI_02 = multiply_columns(df, Year, Rates)
```

```python
# FDI INFLOWS (Ammount in Rupee Crores)
FDI_02.style.set_caption("FDI INFLONS (Amount in  Rupee Crores)").format(precision=2)
```

**Unpivoting DataFrame from wide to long Fromat**

```python
# Unpivoting  melt  DataFrame :
melt = pd.melt(FDI_InUSD, id_vars= Sectors, value_vars= Year, var_name='Year', value_name= 'FDI(US$ Million)', ignore_index=True)
melt

```

```python
# unpivoting melt01 DataFrame :
melt01= pd.melt(FDI_02, id_vars= Sectors, value_vars= Year, var_name='Year', value_name= 'FDI(Rupee Crores)', ignore_index=True)
melt01=round(melt01)
melt01
```

```python
# meringing the  FDI(US$ Million) columns of melt Datafreme into melt01 Dataframe
Merged=melt01.merge(melt,how='left')
Merged
```

```python
# Sorting the Sector and Year columns:
Sorted = Merged.sort_values(['Sector','Year'],ignore_index=True)
Sorted
```

```python
print("\nSats for Sector\n",'-'*65,sep='')
print(pd.DataFrame(Sorted.groupby('Sector').describe().loc[:,:]).transpose())
```

```python
# Repalcing some Long values of Sector Column to Short form
replacements = {
    "CONSTRUCTION DEVELOPMENT: Townships, housing, built-up infrastructure and construction-develo": "CONSTRUCTION DEVELOPMENT",
    "SERVICES SECTOR (Fin., Banking, Insurance, Non Fin/Business, Outsourcing, R&D, Courier, Tech. Testi": "SERVICES SECTOR",
    "TEA AND COFFEE (PROCESSING & WAREHOUSING COFFEE & RUBBER)": "TEA AND COFFEE"
}

Sorted = Sorted[['Sector', 'FDI(Rupee Crores)', 'FDI(US$ Million)']].replace({"Sector": replacements})
```

**Sector- wise Total FDI 2000-2017:**

```python
Sectorwise_fdi =Sorted.groupby('Sector').sum()
Sectorwise_fdi.sort_values(by='FDI(US$ Million)',ascending =False)
```

**VISULAZATION**

```python
Sectorwise_fdi.plot(kind='bar',y='FDI(Rupee Crores)',figsize =(25,7), legend= True, title='SECTOR_WISE FDIINFLOWS',ylabel='FDI')
```

```python
Sectorwise_fdi.plot(kind='bar',y='FDI(US$ Million)',figsize =(25,7), legend= True, title='SECTOR_WISE FDIINFLOWS',ylabel='FDI')
```

**BEST & WORST PERFORMANCE SECTORS**

**#Top 10 and bottom 10 sectors**



```python
Top_10_Sectors = Sectorwise_fdi.nlargest (10, ['FDI(Rupee Crores)'])
```

```python
#Calculating percentage-wise FDI share among top 10 sectors and among all sectors

Total_fdi=round(melt01['FDI(Rupee Crores)'].sum(),2)
Sum = Top_10_Sectors['FDI(Rupee Crores)'].sum()
Top_10_Sectors ['In %age'] = round(Top_10_Sectors['FDI(Rupee Crores)']/Sum*100,2)
Top_10_Sectors['%age to Total Inflows'] = round((Top_10_Sectors['FDI(Rupee Crores)']/Total_fdi)*100,2)
Top_10_Sectors
```

```python
##Creating bar chart to visualise Total FDI inflow in top 10 sectors using Matplotlib



plt.figure(figsize=(15,5))
plt.barh (Top_10_Sectors.index, Top_10_Sectors ['FDI(Rupee Crores)'])
plt.title('SECTORS ATTRACTING HIGHEST FDI INFLOWS')
plt.xlabel('FDI(Rupee Crores)')
plt.ylabel('Top Sectors')
plt.show()
```

```python
#Creating pie chart to visualise percentage share of FDI among top 10 sectors using Matplotlib
plt.figure(figsize=(20,8))
plt.pie(Top_10_Sectors['FDI(Rupee Crores)'], labels =Top_10_Sectors.index, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.title('SHARE AMONG TOP 10 SECTORS ATTRACTING HIGHEST FDI INFLOWS')
plt.show()
```

**From the above Chart, we can understand that Service Sector Managed to Attract highest FDI which was 316347.59Cr greater than any other Sector and among top 10 Sectors it has 27.7% share and among all it has 17.65%**

**BOTTOM 5 SECTORS**

```python
#Calculating share among Bottom sectors and as a whole

Bottom_5_Sectors =Sectorwise_fdi.nsmallest (5, ['FDI(Rupee Crores)'])
Sum = Bottom_5_Sectors['FDI(Rupee Crores)'].sum()
Bottom_5_Sectors['In %age'] = round(Bottom_5_Sectors ['FDI(Rupee Crores)']/Sum*100,2)
Bottom_5_Sectors['%age to Total Inflows'] = round((Bottom_5_Sectors ['FDI(Rupee Crores)']/Total_fdi)*100,3)
Bottom_5_Sectors
```


**Year-wise Details:**

Year-wise FDI Inflow

```python
#Creating Dataframe

melt02 = melt01[['Year', 'FDI(Rupee Crores)']]
melt02=round(melt02.groupby('Year').sum(),2)
 #reating new column of % growth over previous year
melt02['% growth over previous year'] = round(melt02.pct_change()*100,2)
print('\n'+"*"*8+"Details on Variation of FDI INFLOW Year-wise"+'*'*8) #Year-wise FDI Inflow 2
melt02.fillna('-')
```

```python
# Plotting to show Year by year FDI Inflow:

melt02.plot.line(y='FDI(Rupee Crores)',figsize=(10,10))
plt.ylabel('FDI(Rupee Crores)')
plt.title(" FDI Year by year  Inflow")
plt.show()


```

The above graph shows the total amount of FDI inflows in India during the last 17 years i.e. 2000 to 2017.

The FDI inflow from 2000-2001 i.e. 10865.97Cr. in 2001-02 it was 19208.02Cr. It shows the Good result in the FDI inflows in India. Little bit ups and downs in FDI inflows up to 2005-06. but after that great hike in the year 2007-08 and FDI was 98940.57Cr. In 2008-2009 there was a huge investment in FDI in 144390.03Cr. But then there was a downfall in Inflow of FDI in two consecutive years 2009-2010 and 2010-2011, with figures 122558.27 and 97421.29 respectively. We can analysis from the graph that in the year 2011-2012 the inflow of FDI was second highest of last 15 years i.e. 168298.80. Year 2012-13 and 2013-14 the FDI inflow fluctuated from 121984.32 to 147010.90respectively. In last Financial Year i.e.2016-2017 the amount of FDI Inflow were 291608.67Cr which is the highest FDI inflow in last 17 years.
